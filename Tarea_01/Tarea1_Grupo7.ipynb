```python
#Clase para implementar color en los textos enviados por consola
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

#Forma de utilizarlo print(f"{bcolors.OKGREEN}Texto a colorear{bcolors.ENDC}")

# Python program to demonstrate searching operation
# in binary search tree without recursion
class newNode:
 
    # Constructor to create a new node
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

 
# Function to check the given
# key exist or not
# Si la key existe en el arbol, entonces retorna
# las direcciones (Izquierda o Derecha) del camino para encontrarlo 
# y los numeros por los que pasa en el recorrido de busqueda.  
def iterativeSearch(root, key):
    path = []
    numbers_in_path = []
    # Traverse until root reaches
    # to dead end
    while root != None:
         
        # pass right subtree as new tree
        if key > root.data:
            numbers_in_path.append(root.data)
            root = root.right
            path.append("Derecha")
        # pass left subtree as new tree
        elif key < root.data:
            numbers_in_path.append(root.data)
            root = root.left
            path.append("Izquierda")
        else:
            return path, numbers_in_path # if the key is found return path and numbers_in_path
    return False, []
 
# A utility function to insert a
# new Node with given key in BST
def insert(Node, data):
     
    # If the tree is empty, return
    # a new Node
    if Node == None:
        return newNode(data)
 
    # Otherwise, recur down the tree
    if data < Node.data:
        Node.left = insert(Node.left, data)
    elif data > Node.data:
        Node.right = insert(Node.right, data)
 
    # return the (unchanged) Node pointer
    return Node

def compute_height(root):
    """
        Funcion para conocer la altura de un arbol binario
        segun su raiz. En este caso, se toma que un arbol
        con solo la raiz tiene una altura de 1.
        :param root: raiz del arbol binario
        :return: Altura
    """
    if root == None:
        return 0
    
    left_child_height = compute_height(root.left)
    right_child_height = compute_height(root.right)

    return max(left_child_height,right_child_height) + 1

  
def bfs(root):
    """
    Funcion para recorrer el arbol binario por niveles
    Breadth-first search y obtener una lista con los valores de los nodos
    que se recorrieron en el arbol
    :param root: Nodo raiz del arbol a recorrer
    :return: Lista con un arbol completo indicando el recorrido por
    niveles del arbol. Se incluye en la lista los espacios
    vacios en el arbol.
    """

    #Si el arbol es vacio
    if root is None:
      return

    bfs_result = [] #Lista que almacena los nodos recorridos en bfs
    queue = [] #Cola para realizar el recorrido del arbol
    queue.append(root) #Iniciar la cola con la raiz no vacia
    height = compute_height(root) #Calcular la altura para ver el tamaño del arbol

    max_nodos_penultimo_nivel = 0 #Variable para saber cuantos nodos imprimir antes del ultimo nivel

    # Calculo de número posible de nodos del penultimo nivel de un
    # Arbol con altura <height>
    for level in range(height-1):
        max_nodos_penultimo_nivel = max_nodos_penultimo_nivel + 2**(level)
    #print("Cantidad maxima de nodos penultima: "+str(max_nodos_penultimo_nivel))

    nodos_trabajados = 0; #Nodos que han sido recorridos
    while not len(queue) == 0:
        node = queue.pop(0)
        #Contador para saber cuando se llego a la ultima fila
        nodos_trabajados += 1
        #Añadir dato de nodo actual
        bfs_result.append(node.data)

        #Si no se esta en la ultima fila, añadir los datos de los hijos
        #Para procesarlos luego
        if nodos_trabajados <= max_nodos_penultimo_nivel:
            if node.left is not None:
                queue.append(node.left)
            else:
                #si el hijo es vacio, añadir un nodo que representa
                # ese espacio vacio
                queue.append(newNode(' '))
            if node.right is not None:
                queue.append(node.right)
            else:
                #si el hijo es vacio, añadir un nodo que representa
                # ese espacio vacio
                queue.append(newNode(' '))          
              
    return bfs_result

def print_tree(root, numbersToTerm):
    """
    Funcion para imprimir el arbol por consola.
    :param root: Raiz del arbol que se busca imprimir
    :param numbersToTerm: Una lista que incluye los numeros que se quieren
    resaltar en el arbol.
    """
    
    #Verificar si el arbol si tiene una raiz
    if root is None:
      print("No se paso una raiz")
      return


    maxima_altura = compute_height(root)
    nivel_inicial = 1
    nivel_actual = nivel_inicial
    nivel_final = maxima_altura
    factor_espacios = maxima_altura
    nodos_max_nivel_inicial = 2**(nivel_inicial-1)
    nodos_max_nivel_actual = nodos_max_nivel_inicial
    nodos_visitados = 0
    lista_elementos_arbol = bfs(root)
        
    for nivel in range(nivel_final):
        
        espacios_izquierda = 0;
        espacios_centro = 0
        for dummy in range(factor_espacios):
            espacios_izquierda = espacios_izquierda*2 + 1

        for dummy in range(factor_espacios+1):
            espacios_centro = espacios_centro*2 + 1
        
            
        # Tomar los nodos del nivel actual
        nodos_nivel_actual = lista_elementos_arbol[nodos_visitados : nodos_max_nivel_actual]


        nodos_visitados += 2**(nivel_actual-1) 

        #Impresion de los nodos de nivel actual:
        for pos in range(len(nodos_nivel_actual)):
            #Primer nodo en la cadena
            if pos == 0:
                if nodos_nivel_actual[pos] in numbersToTerm:
                    #imprimir en color verde
                    print(bcolors.OKBLUE+" "*espacios_izquierda + str(nodos_nivel_actual[pos])+bcolors.ENDC,end = "")
                else:
                    print(" "*espacios_izquierda + str(nodos_nivel_actual[pos]),end = "") 
            else:
                if nodos_nivel_actual[pos] in numbersToTerm:
                    #imprimir en color verde
                    print(bcolors.OKBLUE+" "*espacios_centro + str(nodos_nivel_actual[pos])+bcolors.ENDC,end = "")
                else:
                    print(" "*espacios_centro + str(nodos_nivel_actual[pos]),end = "") 
        
        # Se pasa al siguiente nivel del arbol
        nivel_actual += 1
        nodos_max_nivel_actual += 2 ** (nivel_actual - 1)
        print()
        #Disminuir el factor de los espacios para el siguiente nivel
        factor_espacios -=1
    
    return
      

# Driver Code
if __name__ == '__main__':
    rootvalue = int(input("Ingrese el valor de la raiz para la generacion de arbol binario: "))
    root = None
    root = insert(root, rootvalue)
    nodeqty = int(input("Cantidad de nodos para la generacion del arbol binario: "))
    for inp in range(nodeqty):
      insert(root, int(input(f"Ingrese el valor ({inp+1}) :")))

    #Input del usuario de que valor desea buscar
    searchTerm = int(input("Ingrese el valor a buscar en el arbol :"))
    pathToTerm, numbersToTerm = iterativeSearch(root, searchTerm)
    numbersToTerm.append(searchTerm)
    if pathToTerm:
      print("Camino para llegar: ", end="")  
      for word in pathToTerm:
        print(word, end=" ")
        
    else:
        print("No se encontró el elemento solicitado")
    print("\nRepresentacion del arbol: ")
    print_tree(root,numbersToTerm)
    
```
