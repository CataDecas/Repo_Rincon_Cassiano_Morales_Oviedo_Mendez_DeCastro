# **Analisis**

El algoritmo realizado se encarga de iterar sobre cada uno de los caracteres de la secuencia del genoma. Cuando se encuentra un caracter igual al primer caracter de la secuencia de target, se inicia un segundo iterador, este último se encarga de iterar desde la posición actual hasta la posición dónde, en caso de encontrar la secuencia debería terminar la misma. 

Para cada uno de los caracteres en este rango, el programa lo compara individualmente con cada uno de los caracteres de la secuencia target. Si al final de la secuencia se encuentra que todos los caracteres son iguales, se indica la posición inicial y final donde se encontró. Si después de llegar al final de la cadena no se encuentra la secuencia target se le indica al usuario. 


Aunque en un primer momento se puede pensar en una complejidad de tipo

$O(n*m)$


Sin embargo, al acercarse a valores muy grandes ambos valores llegan a ser equiparables de modo que la complejidad del algoritmo es


$O(n^2)$

Finalmente, se presenta, en la segunda celda de codigo, un etiquetado de la complejidad de cada linea de codigo. 





```python
target = "AAGGCAGATGGGCTATATAAACGTT"
# Se crea una bandera para verificar si se encontró
final = False
# Iteramos sobre cada caracter dentro de la secuencia

for char in range(len(secuence)):
  flag = True
  # Si el caracter actual es igual al primero del target se itera sobre la 
  # subsecuencia
  if secuence[char] == target[0]:
    try:
      # Se corta la secuencia en la longitud correspondiente (Slicing)
      subseq = secuence[char:char+len(target)]
      # Para cada caracter se compara con el target correspondiente
      for i in range(len(subseq)):
        # Siempre y cuando sean iguales se continua
        if target[i] == subseq[i]:
          continue 
        else:
          flag = False
          break
      # Si al final del ciclo la bandera sigue siendo verdadera se encontró el target 
      if flag == True:
        final = True
        print("Posición inicial: ",char, "\nPosición final:", char+len(target)-1)
    except:
      print("Fallido")

# Si no se encuentra se le avisa al usuario
if final == False:
  print("No encontrado")
```

```python
target = "AAGGCAGATGGGCTATATAAACGTT" # O(1)

# Iteramos sobre cada caracter dentro de la secuencia
for char in range(0, len(secuence)): #O(n)
  flag = True #O(1)
  if secuence[char] == target[0]: # O(1)
    try: # O(1)
      subseq = secuence[char:char+len(target)] # O(1)
      
      for i in range(len(subseq)): #O(n)
        if target[i] == subseq[i]: # O(1)
          continue 
        else:
          flag = False # O(1)
          break
      if flag == True: # O(1)
        print("Posición inicial: ",char, "\nPosición final:", char+len(target))# O(1)
    except:
      print("No encontrado")# O(1)
```
